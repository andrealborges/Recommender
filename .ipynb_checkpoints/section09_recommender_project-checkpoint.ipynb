# Recommender

## Business Understanding

Data from a restaurant review website

Task: Suggests new restaurants to users based on their past restaurant ratings

Task 2: Suggest similar restaurants in each restaurant's pages

```python
# Import
import pandas as pd
import numpy as np
import matplotlib as pl
from matplotlib.pyplot import *
import seaborn as sb

from sklearn.decomposition import TruncatedSVD 
from sklearn.metrics.pairwise import cosine_similarity 
```

## Data Understanding

```python
# Reading data 
df = pd.read_excel('Restaurant_Ratings.xlsx')
df
```

```python
df.describe()
```

- Acess to data of each rating (combination of users and restaurants)
- Max rating = 2; Min rating = 0
- std of o.77 is quite high for the scale

## Data Prep

- For a recommender on user the granularity needs to change to user (1 row 1 user)
- To fill the missing blanks without impacting to much the result the means of the ratings should suffice

```python
# user-item matrix format
data_consumer = df.pivot(index='Consumer_ID', columns='Restaurant_Name', values='Overall_Rating').fillna(df.Overall_Rating.mean())
data_consumer
```

## Modeling

### Collaborative Filtering

**TruncatedSVD model with all the components on centered data**

```python
# centering the data
data_consumer_center = data_consumer - df.Overall_Rating.mean()
data_consumer_center
```

```python
# fit model
svd = TruncatedSVD(n_components=127) 
svd.fit(data_consumer_center) 
```

```python
# Analysing the best num of components 

## matrix U (all lines /  columns = n_components)
U = svd.transform(data_consumer_center) 

U_df = pd.DataFrame(U, index=data_consumer_center.index) 
U_df
```

```python
cumulative = np.cumsum(svd.explained_variance_ratio_).round(4) 
cum_evr = list(zip(range(1, U_df.shape[1] + 1), cumulative)) 
cum_evr
```

```python
cum_evr_df = pd.DataFrame(cum_evr).set_index(0)

cum_evr_df.plot()
```

- Clear that over 40 components the model can explain a satisfing quantity of the variance (>80%)

**Updating the model**

```python
# fit model
svd50 = TruncatedSVD(n_components=50, random_state=42) 
svd50.fit(data_consumer_center)
```

```python
sum(svd50.explained_variance_ratio_) # the model with 50 components has 86% explaned variance
```

***New User***

```python
# vector
vector_len = data_consumer.shape[1]
new_user = np.full(vector_len, np.NaN)

ratings = [2, 2]
positions_to_populate = [40, 60]
new_user[positions_to_populate] = ratings

# DataFram
new_user_df = pd.DataFrame([new_user], columns=data_consumer.columns)

new_user_df.T.dropna()
```

```python
# Data Prep
new_user_preped = new_user_df.fillna(df.Overall_Rating.mean())
new_user_preped = new_user_preped - df.Overall_Rating.mean()

# Model
new_user_tranformed = svd50.transform(new_user_preped)
new_user_tranformed.shape
```

```python
svd50.components_.shape
```

```python
# Predicting 
new_user_recom =  np.dot(new_user_tranformed, svd50.components_)
new_user_recom
```

```python
recom = pd.DataFrame(new_user_recom, columns=data_consumer.columns) 

top_recom = recom.T.sort_values(0, ascending=False) 
top_recom
```

```python
# Displaying the recommendations of the model
already_rated = new_user_df.T.dropna().index 

top_recom = top_recom[~top_recom.index.isin(already_rated)]
top_recom.head(10)
```

- Combining with other analysis

```python
restaurant_details = pd.read_excel('../Data/restaurant_ratings.xlsx', sheet_name=1)
restaurant_details
```

```python
restaurant_details[restaurant_details.Name.isin(new_user_df.T.dropna().index)]
```

User seems to like american cusine with low price

```python
top_recom_merge = top_recom.merge(restaurant_details, how='left', left_on='Restaurant_Name', right_on='Name')
top_recom_merge[(top_recom_merge['Cuisine'] == 'American') | (top_recom_merge['Price'] == 'Low') | (top_recom_merge['Franchise'] == 'Yes')]
```

### Content-Based Filtering

**Prep the data**

```python
restaurants = pd.read_excel('../Data/restaurant_ratings.xlsx', sheet_name=1)
restaurants['Franchise'] = np.where(restaurants['Franchise'] == 'Yes', 1, 0) #Franchise
restaurants.head(10)
```

```python
rest_preping = pd.concat([restaurants, pd.get_dummies(restaurants.Cuisine).astype(int)], axis=1) #Cuisine
rest_preping.head(10)
```

```python
rest_preping['Price'] = np.where(rest_preping.Price == 'High', 1, np.where(rest_preping.Price == 'Medium', 0, -1)) #Price
rest_preping.head(10)
```

```python
rest_preped = rest_preping.set_index(rest_preping.Name)
rest_preped = rest_preped.drop(columns=['Cuisine', 'Restaurant_ID', 'Name'])
rest_preped.head(10)
```

**Cosine similarities between rows**

```python
# calculate all the cosine similarities
cosine_matrix = cosine_similarity(rest_preped) 
cosine_df = pd.DataFrame(cosine_matrix, index=rest_preped.index, columns=rest_preped.index) 
cosine_df
```

```python
def cosine_recommender(item, similarity_df): 

    # check if the input fruit is in the DataFrame 
    if item not in similarity_df.index: 
        return f"{item} is not found in the list." 

    # sort the specified column in descending order 
    sorted_column = similarity_df[[item]].sort_values(by=item, ascending=False) 
    
    #Remove the item
    sorted_column = sorted_column.drop(item, axis=0)
    
    # display the result 
    return sorted_column 
```

```python
top10_recom = cosine_recommender('KFC', cosine_df).head(10)
top10_recom
```

```python
top10_recom.merge(restaurant_details, left_on='Name', right_on='Name')
```

- The recommendations based on the cosine similarity seem to be quite good, the restaurants recommended are really alike
